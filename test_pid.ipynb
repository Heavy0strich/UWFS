```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_points(file):
    ident = open(file)
    x = []
    y = []
    z = []
    while True:
        line = ident.readline()

        if not line:
            break
            
        words = line.split(',', 3)
        
        word_1 = words[0].split(':')
        word_2 = words[1].split(':')
        word_3 = words[2].split(':')

        x.append(float(word_1[2]))
        y.append(float(word_2[1]))
        z.append(float(word_3[1]))

    return x, y, z
```

```python
file_name = 'points.txt'
x, y, z = get_points(file_name)
```

```python
plt.figure(figsize=(10, 10))
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
```
